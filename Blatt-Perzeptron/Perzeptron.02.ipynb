# Perzeptron.02: Logische Funktionen als Perzeptron

```python
#Perzeptron
def perceptron(inputs, weights, threshold):
    s = sum(i*w for i, w in zip(inputs, weights))
    return 1 if s >= threshold else 0


#Logik
def AND(x1, x2):
    return perceptron([x1, x2], [1, 1], 2)

def OR(x1, x2):
    return perceptron([x1, x2], [1, 1], 1)

def NOT(x):
    return perceptron([x], [-1], -0.5)


#Ausgabe
inputs_2 = [(0,0), (0,1), (1,0), (1,1)]
inputs_1 = [0, 1]

print("UND-Funktion:")
for x1, x2 in inputs_2:
    print(f"{x1} AND {x2} = {AND(x1, x2)}")

print("\nODER-Funktion:")
for x1, x2 in inputs_2:
    print(f"{x1} OR {x2} = {OR(x1, x2)}")

print("\nNOT-Funktion:")
for x in inputs_1:
    print(f"NOT {x} = {NOT(x)}")
```

## Warum kann ein Perzeptron kein XOR berechnen?

Ein einzelnes Perzeptron kann nur logische Funktionen berechnen, die **linear separierbar** sind.  
Das bedeutet, dass alle Eingaben, die als 1 klassifiziert werden sollen, auf einer Seite einer einzigen Geraden (Trennebene) liegen müssen, und alle 0-Eingaben auf der anderen.

Bei XOR ist das nicht möglich:

- Die beiden positiven Fälle (0,1) und (1,0) liegen diagonal gegenüber.
- Die beiden negativen Fälle (0,0) und (1,1) liegen ebenfalls diagonal.
- Es gibt keine einzige Gerade, die diese beiden Klassen sauber trennt.
